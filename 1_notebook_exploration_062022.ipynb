# 5: Feature engineering & Exploration des données

<span style="color: #00FFFF"> @ On crée un dataset nommé `data` avec les informations suivantes :
- customer_unique_id
- Récence : Data de la dernière commande;
- Fréquence des commandes : Récurrence d'achat
- Montant : chiffre d'affaire généré
- Satisfaction : Moyenne du review score 
 </span>

```python
data = output.copy()
```

<span style="color: #00FFFF"> On transforme la colonne `order_purchase_last` de type `object` en type `time`. </span>

```python
# On ajoute le mois d'achat du client.
data["order_purchase_month"] = data["order_purchase_timestamp"].dt.month
```

A partir d'un entier et d'une unité de date, 1 jour  :

```python
data_syn = data.groupby(by="customer_unique_id").agg({"order_id":'nunique',
                                                     "payment_value":'sum',
                                                     "review_score":'mean',
                                                    })
data_syn= data_syn.rename(columns ={"order_id":"order_times",
                                    "payment_value":"conso_total",
                                    "review_score":"review_mean",
                                   })

data=pd.merge(data, data_syn, how='left', on="customer_unique_id")
```

```python
data["days_no_purchase"] = data["order_purchase_timestamp"].max()- data["order_purchase_timestamp"]
data["days_no_purchase"]=round(data["days_no_purchase"]/ np.timedelta64(1, 'D'),2)
```

```python
# le nombre de jours depuis le dernier achat
print("la date du premier achat : ", data["order_purchase_timestamp"].min())
print("la date du dernier achat : ", data["order_purchase_timestamp"].max())
```

```python
data.drop_duplicates(subset="customer_unique_id", keep="first", inplace=True)
data.head()
```

```python
cols = ["review_mean", "conso_total","order_times","days_no_purchase"]
for col in cols:
    print("La moyenne de", col, ":", data[col].mean())
```

```python
# supprimer les valeurs manquantes
miss_cols = []
for i in data.columns:
    if data[i].isna().mean()*100!=0:
        miss_cols.append(i)
miss_cols
```

```python
data[miss_cols].isna().mean()
```

```python
for i in miss_cols:
    index_nan = data[data[i].isna()].index
    data.drop(index_nan,axis=0, inplace=True)
```
